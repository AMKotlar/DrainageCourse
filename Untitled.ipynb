```python
import pandas as pd
import matplotlib.pyplot as plt

# read excel file
df = pd.read_excel("progress.xlsx")

# list of column names
columns = list(df.columns)

# create figure and subplots
fig, axes = plt.subplots(nrows=len(columns)-1, ncols=1, figsize=(8,8))

# plot each column except Date
for i, col in enumerate(columns[1:]):
    ax = axes[i]
    ax.plot(df['Date'], df[col])
    ax.set_xlabel('Date')
    ax.set_ylabel(col)

# add title and adjust layout
fig.suptitle('Progress Over Time')
plt.tight_layout()

# show plot
plt.show()
```
