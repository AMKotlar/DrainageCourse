```python
import matplotlib.pyplot as plt
import pandas as pd

# create a figure with one large subplot and three small subplots
fig, (ax, ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=4, figsize=(16,4), sharey=True)

# set the x-axis limits for small subplots
xlims = (30, 130)
x = list(range(1, 366))
y1 = [-90 if i <= 59 else -90 for i in x]
y2 = [-20 if i <= 59 else -90 for i in x]
y3 = [-40 if i <= 59 else -90 for i in x]

y=[y1,y2,y3]

# loop over the names
for i, name in enumerate(names):
    df = pd.read_csv(adrs+"\\"+name+".txt", skiprows=2, header=None, delim_whitespace=True, names=["ID", "time", "Flux"])

    # plot CumSF-4p vs. time
    ax.plot(df['time'], df['Flux'], label=name)
    
    # plot the zoomed-in version in small subplots
    ax_small = [ax1, ax2, ax3][i]
    ax_small.plot(df['time'], df['Flux'],label=name)
    ax_small.set_xlim(xlims)
    ax_right = ax_small.twinx()
    ax_right.step(x, y[i],linestyle='-.', where='post', color ="black")
    ax_right.set_ylim(-100,0)


# set the x-axis label for the large subplot
ax.set_xlabel('day', fontsize=14)
ax_right.set_ylabel('drainage depth[cm]', fontsize=14)

# set the x-axis label for small subplots
ax1.set_xlabel('day', fontsize=12)
ax2.set_xlabel('day', fontsize=12)
ax3.set_xlabel('day', fontsize=12)
ax.set_xlabel('day', fontsize=12)
ax.set_ylabel('Seepage Flux [cm2]', fontsize=14)
ax1.legend()
ax2.legend()
ax3.legend()
ax.legend()

# set the y-axis label for all subplots
```
